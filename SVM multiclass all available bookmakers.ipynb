```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
sheet2018 = pd.ExcelFile("D:/Bets/eurodata/all-euro-data-2018-2019.xlsx")
```

```python
sheet_count = 0
df2018 = pd.read_excel(sheet2018, sheet_name = sheet_count)
sheet_count += 1
for sheet_count in range(1,22):
    df2018 = df2018.append(pd.read_excel(sheet2018, sheet_name = sheet_count),ignore_index=True,sort=False) 
```

```python
df2018['FTR'] = df2018['FTR'].replace('H',0)
df2018['FTR'] = df2018['FTR'].replace('D',1)
df2018['FTR'] = df2018['FTR'].replace('A',2)
df2018
```

```python
#drop columns of match statistics from dataframe -> it's data collected after the event: no predictive value
df = df2018.drop(['Div','Date','HomeTeam','AwayTeam','FTHG','FTAG','HTHG','HTAG','HTR','Referee','HS','AS','HST','AST','HF',
                  'AF','HC','AC','HY','AY','HR','AR','BbOU','Bb1X2','BbMxH','BbAvH','BbMxD','BbAvD','BbMxA','BbAvA',
                  'BbMx>2.5','BbAv>2.5','BbMx<2.5','BbAv<2.5','BbAH','BbAHh','BbMxAHH','BbAvAHH','BbMxAHA','BbAvAHA','HFKC',
                  'AFKC'],axis=1)
```

```python
#drop rows with NaN values
df = df.dropna()
```

```python
target = df['FTR']
```

```python
df = df.drop(['FTR'],axis=1)
```

```python
#data normalization
from sklearn import preprocessing
normalized_data = preprocessing.normalize(df)
normalized_data = normalized_data.astype('float')
```

```python
#data standardization
standardized_data = preprocessing.scale(df)
standardized_data = standardized_data.astype('float')
```

```python
from sklearn import model_selection
from sklearn.svm import SVC
name = 'svm'
results = []
scoring = 'accuracy'
seed = 7
kfold = model_selection.KFold(n_splits=10, random_state=seed)
model = SVC(kernel = 'rbf', gamma = 'scale', shrinking = True, class_weight = 'balanced', decision_function_shape = 'ovo')
cv_results = model_selection.cross_val_score(model, df, target, cv=kfold, scoring=scoring)
results.append(cv_results)
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)
```

```python
model.fit(df,target)
scores = model_selection.cross_val_score(model,df,target,cv=seed,scoring=scoring)
scores.mean()
```
